```python
#Bubble sort

def bubble_sort(list_):
    n = len(list_)
    for i in range(n-1):
        for j in range(n-i-1):
            if list_[j] > list_[j+1]:
                list_[j], list_[j+1] = list_[j+1], list_[j]
    return list_
# Example usage
my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = bubble_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
# prompt: Quick sort

def quick_sort(list_):
    if len(list_) <= 1:
        return list_

    pivot = list_[len(list_) // 2]
    left = [x for x in list_ if x < pivot]
    middle = [x for x in list_ if x == pivot]
    right = [x for x in list_ if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = quick_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
# prompt: Merge sort

def merge_sort(list_):
    if len(list_) <= 1:
        return list_

    mid = len(list_) // 2
    left_half = list_[:mid]
    right_half = list_[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = merge_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
# prompt: Insertion sort

def insertion_sort(list_):
    for i in range(1, len(list_)):
        key = list_[i]
        j = i - 1
        while j >= 0 and key < list_[j]:
            list_[j + 1] = list_[j]
            j -= 1
        list_[j + 1] = key
    return list_

my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = insertion_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
# prompt: Selection sort

def selection_sort(list_):
    n = len(list_)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if list_[j] < list_[min_index]:
                min_index = j
        list_[i], list_[min_index] = list_[min_index], list_[i]
    return list_

my_list = [64, 34, 25, 12, 22, 11, 90]
sorted_list = selection_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
# prompt: Floyd-Warshall Algorithm

def floyd_warshall(graph):

    n = len(graph)
    dist = [[(0) for _ in range(n)] for _ in range(n)]

    # Initialize the distance matrix with the given graph
    for i in range(n):
        for j in range(n):
            dist[i][j] = graph[i][j]

    # Iterate through all possible intermediate vertices
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    return dist
```

```python
# prompt: Dijkstra's Algorithm

import sys

def dijkstra(graph, start):
    num_vertices = len(graph)
    distances = {vertex: sys.maxsize for vertex in range(num_vertices)}
    distances[start] = 0
    visited = set()

    while len(visited) < num_vertices:
        min_distance = sys.maxsize
        min_vertex = -1
        for vertex in range(num_vertices):
            if vertex not in visited and distances[vertex] < min_distance:
                min_distance = distances[vertex]
                min_vertex = vertex

        if min_vertex == -1:
            break

        visited.add(min_vertex)

        for neighbor in range(num_vertices):
            if graph[min_vertex][neighbor] > 0:
                new_distance = distances[min_vertex] + graph[min_vertex][neighbor]
                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance
    return distances

# Example usage:
graph = [
    [0, 4, 0, 0, 0, 0, 0, 8, 0],
    [4, 0, 8, 0, 0, 0, 0, 11, 0],
    [0, 8, 0, 7, 0, 4, 0, 0, 2],
    [0, 0, 7, 0, 9, 14, 0, 0, 0],
    [0, 0, 0, 9, 0, 10, 0, 0, 0],
    [0, 0, 4, 14, 10, 0, 2, 0, 0],
    [0, 0, 0, 0, 0, 2, 0, 1, 6],
    [8, 11, 0, 0, 0, 0, 1, 0, 7],
    [0, 0, 2, 0, 0, 0, 6, 7, 0]
]

start_vertex = 0
distances = dijkstra(graph, start_vertex)
print(f"Shortest distances from vertex {start_vertex}: {distances}")
```

```python
# prompt: Prim's Algorithm

import sys

def prims_algorithm(graph, start_node):
    num_vertices = len(graph)
    key = [sys.maxsize] * num_vertices
    parent = [None] * num_vertices
    mst_set = [False] * num_vertices

    key[start_node] = 0
    parent[start_node] = -1  # Mark the starting node as its own parent

    for _ in range(num_vertices):
        min_key = sys.maxsize
        min_index = -1

        for v in range(num_vertices):
            if key[v] < min_key and not mst_set[v]:
                min_key = key[v]
                min_index = v

        if min_index == -1:
            break

        mst_set[min_index] = True

        for v in range(num_vertices):
            if graph[min_index][v] > 0 and not mst_set[v] and graph[min_index][v] < key[v]:
                key[v] = graph[min_index][v]
                parent[v] = min_index

    return parent, key


# Example usage (using the graph from the Dijkstra's example):
graph = [
    [0, 4, 0, 0, 0, 0, 0, 8, 0],
    [4, 0, 8, 0, 0, 0, 0, 11, 0],
    [0, 8, 0, 7, 0, 4, 0, 0, 2],
    [0, 0, 7, 0, 9, 14, 0, 0, 0],
    [0, 0, 0, 9, 0, 10, 0, 0, 0],
    [0, 0, 4, 14, 10, 0, 2, 0, 0],
    [0, 0, 0, 0, 0, 2, 0, 1, 6],
    [8, 11, 0, 0, 0, 0, 1, 0, 7],
    [0, 0, 2, 0, 0, 0, 6, 7, 0]
]

start_node = 0
parent, key = prims_algorithm(graph, start_node)

print("Edges in the Minimum Spanning Tree:")
for i in range(1, len(graph)):
    print(f"{parent[i]} - {i} \t Weight: {graph[i][parent[i]]}")
```

```python
# prompt: Kruskal's Algorithm

def kruskal_algorithm(graph):
    parent = {}
    rank = {}

    def find(i):
        if parent[i] == i:
            return i
        return find(parent[i])

    def union(i, j):
        root_i = find(i)
        root_j = find(j)
        if root_i != root_j:
            if rank[root_i] < rank[root_j]:
                parent[root_i] = root_j
            elif rank[root_i] > rank[root_j]:
                parent[root_j] = root_i
            else:
                parent[root_j] = root_i
                rank[root_i] += 1
            return True
        return False

    edges = []
    for i in range(len(graph)):
        for j in range(i + 1, len(graph[i])):
            if graph[i][j] != 0:
                edges.append((i, j, graph[i][j]))

    edges.sort(key=lambda x: x[2]) # Sort by weight

    for i in range(len(graph)):
        parent[i] = i
        rank[i] = 0

    minimum_spanning_tree = []
    for u, v, weight in edges:
        if union(u, v):
            minimum_spanning_tree.append((u, v, weight))

    return minimum_spanning_tree

# Example usage (using the graph from the Dijkstra's example):
graph = [
    [0, 4, 0, 0, 0, 0, 0, 8, 0],
    [4, 0, 8, 0, 0, 0, 0, 11, 0],
    [0, 8, 0, 7, 0, 4, 0, 0, 2],
    [0, 0, 7, 0, 9, 14, 0, 0, 0],
    [0, 0, 0, 9, 0, 10, 0, 0, 0],
    [0, 0, 4, 14, 10, 0, 2, 0, 0],
    [0, 0, 0, 0, 0, 2, 0, 1, 6],
    [8, 11, 0, 0, 0, 0, 1, 0, 7],
    [0, 0, 2, 0, 0, 0, 6, 7, 0]
]

mst = kruskal_algorithm(graph)
print("Edges in the Minimum Spanning Tree (Kruskal's Algorithm):")
for u, v, weight in mst:
    print(f"{u} - {v} \t Weight: {weight}")
```

```python
# prompt: Depth-First Search (DFS)

def depth_first_search(graph, start_node):
    visited = set()
    stack = [start_node]
    traversal = []

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            traversal.append(node)
            # Add neighbors to the stack in reverse order for proper DFS
            stack.extend(reversed(graph[node]))
    return traversal

# Example graph represented as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

start_node = 'A'
dfs_traversal = depth_first_search(graph, start_node)
print(f"Depth-First Search traversal starting from {start_node}: {dfs_traversal}")
```

```python
# prompt: Breadth-First Search (BFS)

from collections import deque

def breadth_first_search(graph, start_node):
    visited = set()
    queue = deque([start_node])
    traversal = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            traversal.append(node)
            queue.extend(neighbor for neighbor in graph[node] if neighbor not in visited)
    return traversal

# Example graph represented as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

start_node = 'A'
bfs_traversal = breadth_first_search(graph, start_node)
print(f"Breadth-First Search traversal starting from {start_node}: {bfs_traversal}")
```

```python
# prompt: Strassen's Algorithm

def strassen_matrix_multiplication(A, B):
    """
    Performs matrix multiplication using Strassen's algorithm.

    Args:
        A: A square matrix of size 2^n x 2^n.
        B: A square matrix of size 2^n x 2^n.

    Returns:
        The resulting matrix C = A * B.
    """
    n = len(A)
    if n == 1:
        return [[A[0][0] * B[0][0]]]

    # Split matrices into submatrices
    mid = n // 2
    a11 = [row[:mid] for row in A[:mid]]
    a12 = [row[mid:] for row in A[:mid]]
    a21 = [row[:mid] for row in A[mid:]]
    a22 = [row[mid:] for row in A[mid:]]

    b11 = [row[:mid] for row in B[:mid]]
    b12 = [row[mid:] for row in B[:mid]]
    b21 = [row[:mid] for row in B[mid:]]
    b22 = [row[mid:] for row in B[mid:]]

    # Recursive calls to compute intermediate matrices
    m1 = strassen_matrix_multiplication(add_matrices(a11, a22), add_matrices(b11, b22))
    m2 = strassen_matrix_multiplication(add_matrices(a21, a22), b11)
    m3 = strassen_matrix_multiplication(a11, subtract_matrices(b12, b22))
    m4 = strassen_matrix_multiplication(a22, subtract_matrices(b21, b11))
    m5 = strassen_matrix_multiplication(add_matrices(a11, a12), b22)
    m6 = strassen_matrix_multiplication(subtract_matrices(a21, a11), add_matrices(b11, b12))
    m7 = strassen_matrix_multiplication(subtract_matrices(a12, a22), add_matrices(b21, b22))

    # Combine intermediate matrices to form the resulting matrix
    c11 = add_matrices(subtract_matrices(add_matrices(m1, m4), m5), m7)
    c12 = add_matrices(m3, m5)
    c21 = add_matrices(m2, m4)
    c22 = add_matrices(subtract_matrices(add_matrices(m1, m3), m2), m6)


    # Combine submatrices into the final result
    C = []
    for i in range(mid):
        C.append(c11[i] + c12[i])
    for i in range(mid):
        C.append(c21[i] + c22[i])

    return C


def add_matrices(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            C[i][j] = A[i][j] + B[i][j]
    return C


def subtract_matrices(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            C[i][j] = A[i][j] - B[i][j]
    return C
```

```python
# prompt: Hopcroft-Karp Algorithm

def hopcroft_karp(graph):
    """
    Hopcroft-Karp algorithm for finding a maximum matching in a bipartite graph.

    Args:
        graph: A dictionary representing the bipartite graph where keys are nodes in the left partition
               and values are lists of nodes in the right partition that are connected to the key node.

    Returns:
        A dictionary representing the maximum matching, where keys are nodes in the left partition
        and values are their matched nodes in the right partition.  Nodes not in the matching will not appear
        as keys.
    """
    matching = {}
    while True:
        queue = deque()
        distances = {}

        for node in graph:
            if node not in matching:
                distances[node] = 0
                queue.append(node)
            else:
                distances[node] = sys.maxsize

        distances[-1] = sys.maxsize  # Sentinel node for unmatched right nodes

        while queue:
            u = queue.popleft()

            if distances[u] < distances[-1]:
                for v in graph[u]:
                    if distances[matching.get(v, -1)] == sys.maxsize:
                        distances[matching.get(v, -1)] = distances[u] + 1
                        queue.append(matching.get(v, -1))

        if distances[-1] == sys.maxsize:
            break

        def dfs(u):
            if u != -1:
                for v in graph[u]:
                    if distances[matching.get(v, -1)] == distances[u] + 1:
                        if dfs(matching.get(v, -1)):
                            matching[u] = v
                            matching[v] = u
                            return True
                distances[u] = sys.maxsize
                return False
            return True

        for node in graph:
            if node not in matching and dfs(node):
              pass # We've made a change, which is detected by the while loop condition

    # Clean up the matching, removing right side nodes as keys
    final_matching = {}
    for key, value in matching.items():
      if key in graph:
        final_matching[key] = value

    return final_matching
```

```python
# prompt: Ford-Fulkerson

def ford_fulkerson(graph, source, sink):
    """
    Ford-Fulkerson algorithm to find the maximum flow in a graph.

    Args:
        graph: A dictionary representing the graph where keys are nodes and values are dictionaries of neighbors
               with associated capacities.
        source: The source node.
        sink: The sink node.

    Returns:
        The maximum flow from source to sink.
    """

    flow = 0
    while True:
        queue = deque([(source, float('inf'))])  # (node, path_capacity)
        path = {source: None}

        while queue:
            u, path_capacity = queue.popleft()
            if u == sink:
                break

            for v, capacity in graph[u].items():
                residual_capacity = capacity - graph.get(v, {}).get(u,0)
                if v not in path and residual_capacity > 0:
                    path[v] = u
                    queue.append((v, min(path_capacity, residual_capacity)))

        if sink not in path:  # No augmenting path found
            break

        augmenting_flow = path_capacity

        # Update the flow along the path
        v = sink
        while v != source:
            u = path[v]
            if v not in graph[u]:
              graph[u][v] = 0
            graph[u][v] += augmenting_flow
            if u not in graph[v]:
              graph[v][u] = 0
            graph[v][u] -= augmenting_flow
            v = u

        flow += augmenting_flow

    return flow

# Example Usage
graph = {
    's': {'a': 16, 'b': 13},
    'a': {'c': 12, 'd': 4},
    'b': {'c': 9, 'd': 14},
    'c': {'t': 20},
    'd': {'t': 4}
}
source = 's'
sink = 't'
max_flow = ford_fulkerson(graph, source, sink)
print(f"Maximum Flow from {source} to {sink}: {max_flow}")
```
